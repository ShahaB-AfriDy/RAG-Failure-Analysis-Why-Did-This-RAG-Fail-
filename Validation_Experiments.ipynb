### Importing Required Libraries

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os
import json
```

#### Load environment variables

```python
load_dotenv()
google_api_key = os.getenv("GOOGLE_API_KEY")
```

##### Initializing the Language Model (LLM Gemini)

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite",
    google_api_key=google_api_key,
    temperature=0.9) ## change for different responses
```

### Questions and corresponding context

```python
qa_data = [
    {
        "question": "When will the fall semester begin in 2024?",
        "context": """The fall semester began on September 1, 2023.
The fall semester will begin on September 3, 2024."""
    },
    {
        "question": "Do students need 24 credits to graduate?",
        "context": """Students enrolled before 2023 must complete at least 24 credits to graduate.
Students enrolled in 2023 or later must complete 30 credits to graduate.
Credit requirements differ by enrollment year."""
    },
    {
        "question": "A student enrolled in 2022  how many credits are required?",
        "context": """Students enrolled before 2023 must complete at least 24 credits to graduate.
Students enrolled in 2023 or later must complete 30 credits to graduate."""
    },
    {
        "question": "What changed in thesis submission rules after 2022?",
        "context": """Before 2023, graduate students were required to submit their thesis proposal by March 1.
Starting in 2023, graduate students must submit their thesis proposal by April 1.
Late submissions are accepted with advisor approval."""
    },
    {
        "question": "Which graduation rule applies to a student admitted in early 2023?",
        "context": """Students enrolled before 2023 follow the old graduation rule requiring 24 credits.
Students enrolled in 2023 or later follow the updated graduation rule requiring 30 credits."""
    }
]
```

#### Iterate through all questions and query the LLM

```python
results = []

for item in qa_data:
    prompt = f"""
Answer the question using only the context below.

Context:
{item['context']}

Question:
{item['question']}
"""
    response = llm.invoke(prompt)
    
    results.append({
        "question": item['question'],
        "context": item['context'],
        "answer": response.content
    })

base_path = os.getcwd()
output_path = os.path.join(base_path, "outputs")
os.makedirs(output_path, exist_ok=True)

output_file = os.path.join(output_path, "rag_experiment_results-temp-0.9.json")
with open(output_file, "w") as f:
    json.dump(results, f, indent=4)

print(f"All results saved to {output_file}")
```
